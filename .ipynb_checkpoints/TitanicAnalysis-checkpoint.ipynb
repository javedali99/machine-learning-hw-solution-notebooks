```python
#@title Clean and pre-process data set
#According to the pairplot, the Age and Cabin column have lots of missing values. 
#We get the median of data for the Age and Fare columns, so that we can 
# do exploratory analysis with the Age and Fare data.

# Fill missing values
train['Fare'] = train['Fare'].fillna(train['Fare'].dropna().median())
train['Age'] = train['Age'].fillna(train['Age'].dropna().median())

# We replace the missing values in the Embarked columnn with the label 'S'
train['Embarked']=train['Embarked'].fillna('S')

# To use the sex data, we change the type of the data from categorical to numerical
train.loc[train['Sex']=='male','Sex']=0
train.loc[train['Sex']=='female','Sex']=1

# To use the Embarked data, we also change the type of the data from categorical to numerical
train.loc[train['Embarked']=='S','Embarked']=0
train.loc[train['Embarked']=='C','Embarked']=1
train.loc[train['Embarked']=='Q','Embarked']=2
```

```python
#@title Continue to clean and pre-process data set
variable_name = "" #@param {type:"string"}
#The individual names, the cabin and the ticket are probably less informative
#columns for classifying whether or not the individual survived. 
# Thus, we drop unnecessary columns
drop_elements = ['Name','Cabin','Ticket']
train = train.drop(drop_elements, axis=1)
```

<h2>2. Select a set of important features</h2>

<p style="font-family:verdana, font-size: 1em";>For learning purposes, I am going to experiment with using the classes in the sklearn.feature_selection module for feature selection/dimensionality reduction.</p>

<h3>VarianceThreshold Feature Selection </h3>

<p style="font-family:verdana, font-size: 1em";>VarianceThreshold is a simple baseline feature selection/dimensionality reduction approach that filters (that is, removes) features with low variance. In class, we learned that variance of a random variable X provides a measure of how much the value of X deviates from the mean or expected value of X. A small variance indicates that the data points tend to be very close to the mean, and to each other. A high variance indicates that the data points are very spread out from the mean, and from one another.The VarianceThreshold approach removes all features whose variance doesnt meet some threshold.</p>

<p style="font-family:verdana, font-size: 1em";>For the titanic dataset, I will set a threshold of 0.5, so only those features whose variance is lower than 0.5 (i.e. data points are very close to the mean, and to each other) will be removed. Features whose data are all the same or closely similar do not provide meaningful information for a model. Thus, by setting the threshold moderately at 0.5, we remove these less informative features.</p>

</body>
</html>

```python
#@title Apply Variance Threshold feature selection
from sklearn.feature_selection import VarianceThreshold

modelSelection = VarianceThreshold(threshold=0.5)
modelSelection.fit(train)
ix = modelSelection.get_support() 
```

```python
#@title Show selected features from Variance Threshold method
data1 = pd.DataFrame(modelSelection.transform(train), columns = train.columns.values[ix])
data1.head()
```

<h3>SelectKBest Univariate Feature Selection</h3>

<p style="font-family:verdana, font-size: 1em";>Another feature selection/dimensionality reduction approach is Univariate feature selection that selects the best features based on univariate statistical tests. One univariate feature selection is SelectKBest, which removes all but the  highest scoring features.  SelectKBest uses the chi-square univariate statistical test to test the correlation of qualitative independent variables to qualitative dependent variables. The chi-square test assesses whether the relationship between X and Y that we observed in the data is due to a real relationship between X and Y in the population, or if it is something that could have happened just by chance due to sampling variability.</p>

</body>
</html>

```python
#@title Apply SelectKBest univariate feature selection
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

X = train.drop("Survived",axis=1)
y = train["Survived"]

selection = SelectKBest(chi2, k=5) 
selection.fit(X,y)
ix = selection.get_support() 
```

<h2>7) What can you learn from the algorithm comparision and analysis?</h2>

<p style="font-family:verdana, font-size: 1em";>The Decision Tree algorithm is fast, can handle large datasets and can fit linearly inseparable datasets. The algorithm answers sequential questions that sends us down a path of the tree to the answer. The Decision Tree is simpler to interpret and visualize than the Random Forest algorithm.  The disadvantages to the Decision Tree is that it is sensitive to small data variations, resulting in different tree structures. Decision trees are prone to overfitting, particularly when the tree is deep. Decision trees can lead to errors due to bias and variance. The Random Forest is a collection of decision trees where the results are aggregated into a single final result. The Random Forest algorithm can limit overfitting and can minimize errors due to bias and variance. However, because the Random Forest must compute multiple decision trees, the algorithm is computationally more expensive, especially with very large datasets.</p>

<h2>Are decision trees a linear classifier?</h2>

<p style="font-family:verdana, font-size: 1em";>Decision trees are non-linear classifiers. They are often use to classify non-linearly separable data. The decision trees partition the feature space into smaller regions. There is not single predictive formula that holds over the entire feature space.</p>

<h2>(3) What are the weaknesses of decision trees?</h2>

<p style="font-family:verdana, font-size: 1em";>The disadvantages to the Decision Tree is that it is sensitive to small data variations, resulting in different tree structures. Decision trees are prone to overfitting, particularly when the tree is deep. Finally, Decision trees can lead to errors due to bias and variance. </p>

<h2>(4) Is Misclassification errors better than Gini index as the splitting criteria for decision trees?</h2>

<p style="font-family:verdana, font-size: 1em";>Misclassificatin errors are not better than Gini index as the splitting criteria for decision trees. If we have k classes and node t has frequencies p(t) = (p(1), p(2),...p(k)), the Misclassification error is not smooth in p(t) as in the Gini index. For example, it is possible for the Misclassificatin error to remain the same while the Gini index improves.</p>
