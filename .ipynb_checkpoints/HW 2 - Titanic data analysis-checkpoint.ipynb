## 1. Preprocess the Titanic training data

For machine learning algorithms to work, it is necessary to convert the raw data into a clean data set and dataset must be converted to numeric data. We have to encode all the categorical lables to column vectors with binary values.  Missing values in the dataset is an annoying problem. We have to either drop the missing rows or fill them up with a mean or interpolated  values.

```python
def missing_percentage(df):
    total = df.isnull().sum().sort_values(ascending = False)
    percent = round(df.isnull().sum().sort_values(ascending = False)/len(df)*100, 2)
    return pd.concat([total, percent], axis=1, keys=['Total','Percent'])
```

```python
# check the missing values in training dataset
missing_percentage(train_df)
```

```python
# check th emissing values in test dataset
missing_percentage(test_df)
```

We can see from the above results that the `Cabin` and `Age` columns in training and testing data have lots of missing values.

```python
# dropping the columns which are not useful

print("Before dropping the columns: ", train_df.shape, test_df.shape, combine_df[0].shape, combine_df[1].shape)

train_df = train_df.drop(['Ticket', 'Cabin'], axis=1)
test_df = test_df.drop(['Ticket', 'Cabin'], axis=1)
combine_df = [train_df, test_df]

print("After dropping the unuseful columns: ", train_df.shape, test_df.shape, combine_df[0].shape, combine_df[1].shape)
```

### Creating new feature from existing

We want to analyze if `Name` feature can be engineered to extract titles and test correlation between titles and survival, before dropping Name and `PassengerId` features.

```python
for data in combine_df:
    data['Title'] = data.Name.str.extract(' ([A-Za-z]+)\.', expand=False)

pd.crosstab(train_df['Title'], train_df['Sex'])
```

We can replace many titles with a more common name or classify them as `Other`.

```python
for data in combine_df:
    data['Title'] = data['Title'].replace(['Lady', 'Countess','Capt', 'Col',\
 	'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Other')

    data['Title'] = data['Title'].replace('Mlle', 'Miss')
    data['Title'] = data['Title'].replace('Ms', 'Miss')
    data['Title'] = data['Title'].replace('Mme', 'Mrs')
    
train_df[['Title', 'Survived']].groupby(['Title'], as_index=False).mean()
```

```python
# convert the categorical titles to ordinal
title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Other": 5}

for data in combine_df:
    data['Title'] = data['Title'].map(title_mapping)
    data['Title'] = data['Title'].fillna(0)

train_df.head()
```

Create a new feature for `FamilySize` that combines `Parch` and `SibSp` features. Afterwards, we can drop `Parch` and `SibSp` from our dataset

```python
for data in combine_df:
    data['FamilySize'] = data['SibSp'] + data['Parch'] + 1

train_df[['FamilySize', 'Survived']].groupby(['FamilySize'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

We can drop the `Name` feature from training and testing datasets. We also do not need the `PassengerId` feature in the training dataset.

```python
train_df = train_df.drop(['Name', 'PassengerId'], axis=1)
test_df = test_df.drop(['Name'], axis=1)

combine_df = [train_df, test_df]

train_df.shape, test_df.shape
```

Now, we can drop `Parch`, and `SibSp` features.

```python
train_df = train_df.drop(['Parch', 'SibSp'], axis=1)
test_df = test_df.drop(['Parch', 'SibSp'], axis=1)
combine_df = [train_df, test_df]

train_df.head()
```

### Converting categorical features to numerical and filling the missing values

```python
# change the sex data from categorical to numerical
for data in combine_df:
    data['Sex'] = data['Sex'].map( {'female': 1, 'male': 0} ).astype(int)

train_df.head()
```

```python
# filling missing values in the Age and Fare columns using the median values of the data
train_df['Fare'] = train_df['Fare'].fillna(train_df['Fare'].dropna().median())
train_df['Age'] = train_df['Age'].fillna(train_df['Age'].dropna().median())

# filling the missing values in the Embarked columnn with the most common embarked port 'S'
port_freq = train_df['Embarked'].dropna().mode()[0]
train_df['Embarked']= train_df['Embarked'].fillna(port_freq)

train_df[['Embarked', 'Survived']].groupby(['Embarked'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
# convert the embarked data from categorical to numerical
for data in combine_df:
    data['Embarked'] = data['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)

train_df.head()
```

## 2. Select a set of important features

### VarianceThreshold Feature Selection using `sklearn.feature_selection`

`VarianceThreshold` is a simple feature selection approach that filters features with low variance. The variance of a random variable X provides a measure of how much X's value deviates from X's mean value. A slight variation shows that the data points appear to be very close to each other and to the mean value. A high variance means that the data points are quite spread out from the mean and from each other. The `VarianceThreshold` approach removes all features whose variance doesnt meet a threshold. For the given titanic data, we can set a threshold of 0.5, so features having variance value lower than 0.5 will be removed. 

The closely similar or the same data do not provide meaningful information. So, we will remove less informative features by setting the threshold at 0.5.

```python
# selcting important features using VarianceThreshold feature selection method

modelSelect = VarianceThreshold(threshold=0.5)
modelSelect.fit(train_df)

ix = modelSelect.get_support() 
```

```python
# selected features from VarianceThreshold method

data_imp = pd.DataFrame(modelSelect.transform(train_df), columns = train_df.columns.values[ix])
data_imp.head()
```

Using the VarianceThreshold method, the selected features are `Pclass`, `Age`, `Fare`, `Title` and `FamilySize`.

### SelectKBest Univariate Feature Selection

`SelectKBest` univariate feature selection is another feature selection approach, whihc is used to select the best features based on univariate statistical tests. `SelectKBest` removes all features except the highest scoring features. `SelectKBest` uses the chi-square univariate statistical test to test the correlation of qualitative independent variables to qualitative dependent variables. The chi-square test assesses whether the relationship between X and Y that we observed in the data is due to a real relationship between X and Y in the population, or if it is something that could have happened just by chance due to sampling variability.

```python
X = train_df.drop("Survived", axis=1)
y = train_df["Survived"]

modelSelection = SelectKBest(chi2, k=5) 
modelSelection.fit(X,y)
ix = modelSelection.get_support() 

# selected features from SelectKBest univariate feature selection method
data_imp2 = pd.DataFrame(modelSelection.transform(X), columns = X.columns.values[ix]) 
data_imp2.head(n=5)
```

Using the `SelectKBest` univariate feature selection method, the selected features are `Pclass`, `Sex`, `Age`, `Fare`, and `Title`.

### `SelectFromModel` for Logistic Regression

```python
X = train_df.drop("Survived",axis=1)
y = train_df["Survived"]

# Linear Model
lin_model = LogisticRegression()
lin_model.fit(X,y)

model = SelectFromModel(lin_model, prefit=True)
ix = model.get_support() 

# # selected features
data_imp3 = pd.DataFrame(model.transform(X), columns = X.columns.values[ix]) 
data_imp3.head(n=5)
```

Using `SelectFromModel` method, the selected features are `Pclass` and `Sex`.

### Recursive Feature Selection

Given an external estimator that assigns weights to features (e.g., the coefficients of a linear model), the goal of recursive feature elimination (RFE) is to select features by recursively considering smaller and smaller sets of features.

```python
modelSelect2 = RFE(lin_model,n_features_to_select=5)
modelSelect2.fit(X,y)
ix = modelSelect2.get_support() 

# selected features
data_imp4 = pd.DataFrame(modelSelect2.transform(X), columns = X.columns.values[ix]) 
data_imp4.head(n=5) 
```

Using `RecursiveFeatureElimination (RFE)` method, the selected features are `Pclass`, `Sex`, `Embarked`, `Title` and `FamilySize`.

The `RecursiveFeatureElimination (RFE)` method selected `Pclass`, `Sex`, `Embarked`, `Title` and `FamilySize`. `SelectKBest` univariate feature selection method selected `Pclass`, `Sex`, `Age`, `Fare`, and `Title` and the `VarianceThreshold` feature selection method selected `Pclass`, `Age`, `Fare`, `Title` and `FamilySize`.

## 3. Learn a Decision Tree model with the Titanic training data using Gini Index and plot the Decision Tree.

```python
# decision tree model with Titanic training dataset using Gini index

target = train_df['Survived'].values
selected_features = ['Pclass','Sex','Fare', 'Title','Embarked','Age', 'FamilySize']
features = train_df[selected_features].values

# build training and testing datasets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state = 42)

decision_tree = tree.DecisionTreeClassifier(random_state = 1, criterion = 'entropy', min_samples_split = 100)

decision_tree_1 = decision_tree.fit(X_train, y_train)
target_predict = decision_tree_1.predict(X_test)


print("Decision tree score: ", accuracy_score(y_test, target_predict))

print ("MSE    :",mean_squared_error(y_test, target_predict))

print ("R2     :",r2_score(y_test, target_predict))
```

```python
# plot the decision tree

dec_tree = tree.DecisionTreeClassifier(
        random_state = 1,
        max_depth = 5,
        min_samples_split=2
)

dec_tree_1 = dec_tree.fit(features, target)

print("Generalized decision tree score: ", dec_tree_1.score(features,target))

dot_data = tree.export_graphviz(dec_tree_1, feature_names = selected_features, out_file = None)

plot_dc = graphviz.Source(dot_data)

plot_dc.format = "png"
plot_dc.render("decision-tree")

plot_dc
```

## 7. What are your observations and conclusions from the algorithm comparison and analysis?

### Decision Tree
The `Decision Tree` algorithm is fast, can handle large datasets and can fit linearly inseparable datasets. A decision tree is simply a series of sequential decisions made to reach a specific result. Decision trees are much easier to interpret and understand. Since a random forest combines multiple decision trees, it becomes more difficult to interpret. The disadvantages to the `Decision Tree` is that it is sensitive to small data variations, resulting in different tree structures. Decision trees are prone to overfitting, particularly when the tree is deep. Decision trees can lead to errors due to bias and variance. Decision tree gives low prediction accuracy for a dataset as compared to other machine learning algorithms.

### Random Forest
The `Random Forest` is typically an ensemble of `Decesion Tree` classifer, where the results are aggregated into a single final result. It performs better than `Decision Tree`. The Random Forest algorithm can limit overfitting and can minimize errors due to bias and variance. As the Random Forest algorithm combines the output of multiple (randomly created) Decision Trees to generate the final output, it is computationally more expensive, especially with very large datasets. Random Forest has a higher training time than a single decision tree. We should take this into consideration because as we increase the number of trees in a random forest, the time taken to train each of them also increases. That can often be crucial when youre working with a tight deadline in a machine learning project.

Random forest does not rely on the feature importance given by a single decision tree. The decision tree model gives high importance to a particular set of features. But the random forest chooses features randomly during the training process. Therefore, it does not depend highly on any specific set of features. This is a special characteristic of random forest over bagging trees. Therefore, the random forest can generalize over the data in a better way. This randomized feature selection makes random forest much more accurate than a decision tree.


Despite instability and dependency on a particular set of features, decision trees are really helpful because they are easier to interpret and faster to train. Anyone with very little knowledge of data science can also use decision trees to make quick data-driven decisions.
